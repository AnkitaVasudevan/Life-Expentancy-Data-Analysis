```python
import pandas as pd
import numpy as np
from prettytable import PrettyTable
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = df.drop(['Hepatitis B','Polio','Total expenditure',' thinness 5-9 years','Schooling','under-five deaths '],axis=1)
```

```python
#Capitalisation of the categorical data
df['Country'] = df['Country'].str.upper()
df['Status'] = df['Status'].str.upper()
df.head()
```

```python
df = df[df['Year'] >2010]
```

```python
## MEAN and VARIANCE
```

```python
p=PrettyTable()
p.field_names=["Parameter","Mean","Variance"]
```

```python
lfm=df['Life expectancy '].mean()
x=df['Life expectancy '].std()
lfv=x**2
amm=df['Adult Mortality'].mean()
y=df['Adult Mortality'].std()
amv=y**2
idm=df['infant deaths'].mean()
z=df['infant deaths'].std()
idv=z**2
am=df['Alcohol'].mean()
a=df['Alcohol'].std()
av=a**2
pem=df['percentage expenditure'].mean()
b=df['percentage expenditure'].std()
pev=b**2
mm=df['Measles '].mean()
c=df['Measles '].std()
mv=c**2
bm=df[' BMI '].mean()
d=df[' BMI '].std()
bv=d**2
dm=df['Diphtheria '].mean()
e=df['Diphtheria '].std()
dv=e**2
ham=df[' HIV/AIDS'].mean()
f=df[' HIV/AIDS'].std()
hav=f**2
gm=df['GDP'].mean()
g=df['GDP'].std()
gv=g**2
pm=df['Population'].mean()
h=df['Population'].std()
pv=h**2
tm=df[' thinness  10-19 years'].mean()
i=df[' thinness  10-19 years'].std()
tv=i**2
icm=df['Income composition of resources'].mean()
j=df['Income composition of resources'].std()
icv=j**2

p.add_row(["Life expectancy",lfm,lfv])
p.add_row(["Adult mortality",amm,amv])
p.add_row(["Infant deaths",idm,idv])
p.add_row(["Alcohol",am,av])
p.add_row(["Percentage expenditure",pem,pev])
p.add_row(["Measles",mm,mv])
p.add_row(["BMI",bm,bv])
p.add_row(["Diphtheria",dm,dv])
p.add_row(["HIv/AIDS",ham,hav])
p.add_row(["GDP",gm,gv])
p.add_row(["Population",pm,pv])
p.add_row(["Thinness 1-19 Years",tm,tv])
p.add_row(["Income composition of resources",icm,icv])

print(p)
```
